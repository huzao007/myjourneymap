# My Travel History


### As a method to practice, learn and improve on geo-spatial data visualization I wanted to create an interactive map that visualizes all the cities I have travelled to across the world. 


```python
#Import necessary Libraries.

import folium
import pandas as pd
```

```python
cities = pd.read_csv("Mytravelledcities.csv") #Load CSV data to pd

#Clean up some of the data. Some LAT contained comma at the end since it was scraped from external sources. 
#Remove Unnecessary comma
cities['Lat']=cities['Lat'].str.replace(',','')  

#Converting Lat to Float64
cities['Lat'] = pd.to_numeric(cities['Lat'] )
cities['lon'] = pd.to_numeric(cities['lon'] )
```

```python
#load the folium map and assign to dataset
my_map = folium.Map(
    location = [13.133932434766733, 16.103938729508073],
    zoom_start=2)
```

```python
#Setting a marker for first entry 
city = cities.loc[0]
folium.Marker(
    location=[city['Lat'],city['lon']]).add_to(my_map)

```

```python
my_map = folium.Map(
    location = [13.133932434766733, 16.103938729508073],
    zoom_start=2,
    )

#Creating loop to enter all the city on the map:

for _, city in cities.iterrows():
    folium.Marker(
    location=[city['Lat'], city['lon']],
        popup=city['City'],
        tooltip=city['City'],
        
        icon=folium.Icon(color=city['color'], prefix='fa', icon='user')
    ).add_to(my_map)
```

```python
#Creating a function to set a color for home cities 
def select_marker_color(row):
    if row['Type'] == 'Home':
        return 'blue'
    elif row['Type'] == 'Travel':
        return 'gold'
    
```

```python
#Adding color as a column so it can be passed to select color for markers based on 'Type'
cities['color']=cities.apply(select_marker_color, axis=1)
cities.head()
```
